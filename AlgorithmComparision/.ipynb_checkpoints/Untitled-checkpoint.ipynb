```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
kpts=pd.read_csv('Keypoints.csv',header=None)
```

```python
def rearrange(data):
    d={}
    for i,j in enumerate(kpts.iloc[:,0]):
        #for num in kpts.iloc[i,1:]:
        #    d[i]=
        temp=[]
        for num in kpts.iloc[i,1:]:
            temp.append(num)
        d[j]=temp
    return d

def indexDefine(value):
    temp = []
    value=value+"_"
    for i in range(1,11):
        label=value+str(i)
        temp.append(label)
    return temp
```

```python
keypts=pd.DataFrame(rearrange(kpts),index=indexDefine("Image"))
keypts=keypts.sort_values(by=indexDefine("Image"),axis=1,ascending=False)
```
